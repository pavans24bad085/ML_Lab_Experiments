```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score

import kagglehub
```

## Load Dataset

```python
path = kagglehub.dataset_download('sohier/calcofi')
print('Dataset path:', path)

df = pd.read_csv(f'{path}/bottle.csv')
df.head()
```

## Select Features and Target

```python
features = ['Depthm', 'Salnty', 'O2ml_L', 'Lat_Dec', 'Lon_Dec']
target = 'T_degC'

data = df[features + [target]]
data.info()
```

## Handle Missing Values

```python
data = data.fillna(data.median())
data.isnull().sum()
```

## Feature Scaling

```python
X = data[features]
y = data[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

## Linear Regression Model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

## Model Evaluation

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('MSE:', mse)
print('RMSE:', rmse)
print('R2 Score:', r2)
```

## Visualization: Actual vs Predicted

```python
plt.figure()
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Temperature')
plt.ylabel('Predicted Temperature')
plt.title('Actual vs Predicted Water Temperature')
plt.show()
```

## Visualization: Residuals

```python
residuals = y_test - y_pred

plt.figure()
sns.histplot(residuals, bins=30, kde=True)
plt.title('Residual Error Distribution')
plt.xlabel('Residuals')
plt.show()
```

## Regularization: Ridge Regression

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)

print('Ridge RMSE:', np.sqrt(mean_squared_error(y_test, ridge_pred)))
print('Ridge R2:', r2_score(y_test, ridge_pred))
```

## Regularization: Lasso Regression

```python
lasso = Lasso(alpha=0.01)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

print('Lasso RMSE:', np.sqrt(mean_squared_error(y_test, lasso_pred)))
print('Lasso R2:', r2_score(y_test, lasso_pred))
```

## Conclusion
This notebook demonstrated how linear models and regularization techniques can be applied to predict ocean water temperature using the CalCOFI dataset.
