Numerical Features for Correlation Scatter Plot and Heatmaps

```python
num_df = df.select_dtypes(include=["int64", "float64"])
target = "price" 
```

```python
sns.set(style="whitegrid")

plt.figure()
sns.scatterplot(x=df["area"], y=df[target])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 8))
corr = num_df.corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
